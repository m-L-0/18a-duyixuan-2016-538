### 数据分析前操作

```python
import scipy.io as si
data2=si.loadmat("9-train/data2_train.mat")
data3=si.loadmat("9-train/data3_train.mat")
data5=si.loadmat("9-train/data5_train.mat")
data6=si.loadmat("9-train/data6_train.mat")
data8=si.loadmat("9-train/data8_train.mat")
data10=si.loadmat("9-train/data10_train.mat")
data11=si.loadmat("9-train/data11_train.mat")
data12=si.loadmat("9-train/data12_train.mat")
data14=si.loadmat("9-train/data14_train.mat")
```

```python
import numpy as np
labels=np.ones([len(data2['data2_train']),1])
label2=np.where(labels>0,2,0)

labela=np.ones([len(data3['data3_train']),1])
label3=np.where(labela>0,3,0)


labelb=np.ones([len(data5['data5_train']),1])
label5=np.where(labelb>0,5,0)


labelc=np.ones([len(data6['data6_train']),1])
label6=np.where(labelc>0,6,0)


labeld=np.ones([len(data8['data8_train']),1])
label8=np.where(labeld>0,8,0)


labele=np.ones([len(data10['data10_train']),1])
label10=np.where(labele>0,10,0)


labelf=np.ones([len(data11['data11_train']),1])
label11=np.where(labelf>0,11,0)


labelg=np.ones([len(data12['data12_train']),1])
label12=np.where(labelg>0,12,0)


labelh=np.ones([len(data14['data14_train']),1])
label14=np.where(labelh>0,14,0)
dt2=np.c_[data2['data2_train'],label2]
dt3=np.c_[data3['data3_train'],label3]
dt5=np.c_[data5['data5_train'],label5]
dt6=np.c_[data6['data6_train'],label6]
dt8=np.c_[data8['data8_train'],label8]
dt10=np.c_[data10['data10_train'],label10]
dt11=np.c_[data11['data11_train'],label11]
dt12=np.c_[data12['data12_train'],label12]
dt14=np.c_[data14['data14_train'],label14]
```

```python
data_array1 = np.concatenate((dt2,dt3,dt5))
data_array2=np.concatenate((dt6,dt8,dt10))
data_array3=np.concatenate((dt11,dt12,dt14))
data_array=np.concatenate((data_array1,data_array2,data_array3))
```

#### 数值型数据统计
